#### Monthly usage time of learners who have confirmed and are not deleted

```python
%%sql
SELECT COUNT(lastaccess - firstaccess) AS usagetime, 
    EXTRACT (MONTH FROM to_timestamp(firstaccess)) AS month 
        FROM mdl_user 
            WHERE confirmed = 1 AND deleted = 0 
                GROUP BY EXTRACT (MONTH FROM to_timestamp(firstaccess))
```

```python
%%sql
SELECT id, confirmed, deleted, lastaccess - firstaccess AS usagetime, 
EXTRACT(MONTH FROM to_timestamp (firstaccess)) AS Month 
FROM mdl_user 
WHERE confirmed = 1 AND deleted = 0
GROUP BY (EXTRACT(MONTH FROM to_timestamp (firstaccess)))
LIMIT 10;
```

#### Number of quiz submissions by hour of day

```python
%%sql
SELECT Extract(HOUR FROM to_timestamp(timecreated)) AS hour, count(*) 
FROM mdl_logstore_standard_log where action='submitted' AND component like '%quiz%' GROUP BY hour;
```

#### Dedication Time

```python
#only_log = df_log[(df_log.action == "loggedin") | (df_log.action == "loggedout")]
```

```python
user_id =  df_log.userid.unique()
d_times = {}
for user in tqdm(user_id):
    log_user = df_log[df_log.userid == user].sort_values('timecreated')
    d_time = 0
    isLoggedIn = 0
    loggedIn_timecreated = 0
    for i in range(len(log_user)):
        row = log_user.iloc[i,]
        row_next = log_user.iloc[i+1,] if i+1 < len(log_user) else row
        if(row.action == "loggedin"):
            isLoggedIn = 1
            loggedIn_timecreated = row.timecreated
        if( (i+1 == len(log_user)) | ( (row_next.action == "loggedin") & (isLoggedIn == 1) ) ):
            d_time += min(row.timecreated - loggedIn_timecreated, 7200)
            isLoggedIn = 0
    d_times[user] = d_time
```

```python
d_times
```

```python
dedication_time_df = pd.DataFrame( {'userid':list(d_times.keys()), 
                                    'dedication_time':list(d_times.values())})
```

```python
def learner_percent(df, percent):
    tot_len = df.shape[0]
    top = int((tot_len * percent)/100)
    return df.iloc[:top,]
```
